```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for i in range(len(data)):
    data.at[i, 'ip'] = data.at[i, 'ip'][:-3] + str(np.random.randint(0, 256))
```

```python
 def impute_browser(cols):
        Browser = cols[0]
        if pd.isnull(Browser):
                return 'linux'
        else:
                return Browser
       
                
            
```

```python
data['browser'] = data[['browser']].apply(impute_browser, axis=1)
```

```python
sns.boxplot(x='browser', y='size', data=data)
```

```python
 def impute_size(cols):
        Size = cols[0]
        if pd.isnull(Size):
                return str(np.random.randint(0, 200000))
        else:
                return Size
       
```

```python
data['size'] = data[['size']].apply(impute_size, axis=1)
```
