```python
 def impute_browser(cols):
        Browser = cols[0]
        if pd.isnull(Browser):
                return 'linux'
        else:
                return Browser
       
                
            
```

```python
 def impute_size(cols):
        Size = cols[0]
        if pd.isnull(Size):
                return str(np.random.randint(0, 200000))
            
        else:
                return Size
       
```
